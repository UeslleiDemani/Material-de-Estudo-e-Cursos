## Leitura Dataset - Estações Metrô SP

https://www.kaggle.com/thiagodsd/sao-paulo-metro

```python
metro = pd.read_csv('metrosp_stations.csv')
```

```python
metro.head(10)
```

```python
a = zip(metro['lat'], metro['lon'])
```

```python
# Geração da lista com endereços - Lat, Long das estações
ends = list(map(list, zip(metro['lat'], metro['lon'])))
ends
```

```python
# coloca nome das estações em uma lista
nomes = list(metro['name'])
nomes
```

```python
# estrutura de repetição para carga dos marcadores das estações de metrô
for gps, nome in zip(ends, nomes):
    folium.Marker(location=gps, tooltip=nome).add_to(mapa)
```
