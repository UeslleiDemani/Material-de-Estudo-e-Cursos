# APIs do matplotlib

Matplotlib permite criar gráficos de duas maneiras, usando o módulo pyplot ou usando a interface orientada a objetos. Os comandos de plot do pyplot são similares ao MATLAB e com poucos comandos é possível criar um gráfico sem especificar cada componente da figura. Por outro lado, usando a opção orientada a objetos é possível acessar e modificar cada componente da figura de maneira mais direta. Para uma comparação mais detalhada das duas apis: [link](https://matplotlib.org/matplotblog/posts/pyplot-vs-object-oriented-interface/)

## Gráfico com múltiplas linhas. Adicionando Títulos, legendas e eixos

```python
##### Having more than 1 line - Adding Title, legends and x/y tickers

# Geração dos valores a serem apresentados
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,2,4,6,5])

# tamanho do gráfico
plt.figure(figsize=(10,10))
# plot com definição de cor e estilo da linha 
plt.plot(a, color = 'green', linestyle = '-.', label="Aluguel 2020")
# plot com definição de largura da linha 
plt.plot(b, linewidth=5., label="Valor Medio Aluguel")
# plot com definição de cor e estilo da linha 
plt.plot(c, color = 'red', linestyle = '--', label="Aluguel 2019")
# título do gráfico
plt.title("Resumo - Valores Aluguel por Mes", fontsize=20, fontweight="bold")
plt.title("Resumo - Valores Aluguel", fontsize=20, fontweight="bold")

# legenda do gráfico - no alto à esquerda
plt.legend(loc="upper left", fontsize=15)
# valores no eixo
plt.xticks(range(0,len(a)),('Jan','Fev','Mar','Abr','Mai'))
# descrição eixo X
plt.xlabel("Meses")
# descrição eixo Y
plt.ylabel("Valor 1000$")
plt.show()
```

## Subplots - múltiplos gráficos

```python
# Gração dos Dados
categoria = ['Conta A','Conta B','Conta C']
faturamento = np.array([300, 450, 700])

# Plotting it
plt.figure(figsize=(10,5))

# Barra
plt.subplot(131)
plt.bar(range(0,3), faturamento)
plt.xticks(range(0,3),categoria)

# Linha
plt.subplot(132)
plt.plot(range(0,3), faturamento, c="red")
plt.xticks(range(0,3),categoria)

# Dispersão
plt.subplot(133)
plt.scatter(range(0,3), faturamento, c="green")
plt.xticks(range(0,3),categoria)
plt.show()
```

## Gráficos Linhas - Base de Dados de Marketing

```python
# Leitura dados - CSV
raw_data = pd.read_csv("Marketing Raw Data.csv")
# 'shape' do dataset
print(raw_data.shape)
# 'head' do dataset
raw_data.head()
```

```python
# Gráfico de Linhas - simples
plt.plot(raw_data["Visitors"])
plt.show()
```

```python
# seleciona os últimos 10 registros do dataset
a = raw_data.tail(10).reset_index()
plt.figure(figsize=(15,10))
# plota os valores e de visitantes e faturamento
plt.plot(a["Visitors"], label="Quantidade de Visitantes", linewidth=5.)
plt.plot(a["Revenue"], linestyle = "-.")

plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitantes & Faturamento por Dia", fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Data")
plt.show()
```

## Combinações de Gráficos

```python
plt.figure(figsize=(15,10))

# plota os valores e de visitantes e faturamento
plt.bar(range(0,len(a["Visitors"])), a["Visitors"], label="Quantidade de Visitantes", linewidth=5., color = "green")
plt.plot(a["Revenue"], linestyle = "-.")

plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitantes & Faturamento por Dia", fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Data")
plt.show()
```

```python
plt.figure(figsize=(15,10))

# Barras empilhadas
plt.bar(range(0,len(a["Revenue"])), a["Revenue"], label="Faturamento", linewidth=5., color = "green")
plt.bar(range(0,len(a["Visitors"])), a["Marketing Spend"], label="Gasto Marketing", linewidth=5., color = "red")

plt.plot(a["Visitors"], linestyle = "-.")

plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitantes & Faturamento por Dia", fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Data")
plt.show()
```

## 6. Gráficos de Dispersão

```python
plt.figure(figsize=(15,10))
plt.scatter(raw_data["Revenue"], raw_data["Marketing Spend"], c="red", marker='+')
plt.title("Gasto Marketing x Faturamento", fontsize = 20)
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Faturamento")
plt.ylabel("Gasto Marketing")

plt.show()
```

```python
groups.describe()
```

```python
# Utilizando grupos com Gráficos de Dispersão - Promoção de Marketing

groups = raw_data[["Promo","Revenue","Marketing Spend"]].groupby("Promo")

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo)

ax.legend()
plt.title("Gasto de Marketing & Faturamento por Promo", fontsize = 20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```

```python
# Utilizando grupos com Gráficos de Dispersão - Dia da Semana

groups = raw_data[["Day_Name","Revenue","Marketing Spend"]].groupby("Day_Name")
groups

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo)

ax.legend()
plt.title("Relationship between M. Spend & Revenue by Day_Name", fontsize = 20)
plt.xlabel("Faturamento")
plt.ylabel("Gasto Marketing")

plt.show()
```

```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Leitura - Dados Marketing

```python
raw_data = pd.read_csv("Marketing Raw Data.csv")
raw_data.head()
```

## Gráficos Linha + Intervalo de Confiança

```python
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data)
```

```python
# Podemos quebrar o gráfico pelo 'Promo'

ax = sns.lineplot(x='Week_ID', y='Revenue', hue = 'Promo', data = raw_data)
```

## Gráficos de Barras

```python
# o intervalo de confiança é impresso por default
ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data)
#raw_data[['Month_ID', 'Revenue']].groupby('Month_ID', as_index = False).agg({'Revenue':'mean'})
```

```python
# Vamos remover o intervalo de confiança
ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data, ci=False)
```

```python
# Faturamento por Mês - quebrado por Promo
plt.figure(figsize=(15,10))
ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data, ci=False, hue = 'Promo')
```

```python
# Faturamento por Ano - quebrado por Promo - Barras Horizontais
ax = sns.barplot(x="Revenue", y="Year", ci=False, hue = 'Promo', orient = 'h', data=raw_data)
```

## Histogramas

```python
# Distribuição do Faturamento
x = raw_data['Revenue'].values
sns.distplot(x, color = 'blue');
```

```python
x = raw_data['Revenue'].values
sns.histplot(x, color = 'blue');
```

```python
# Adicionando a média no gráfico anterior

x = raw_data['Revenue'].values

sns.distplot(x, color = 'blue');

# Cálculo da Média
mean = raw_data['Revenue'].mean()

plt.axvline(mean, 0,1, color = 'red')
```

## Gráficos Dispersão

```python
# Gráfico Dispersão básico
ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=raw_data, color = 'green')
```

```python
# Quebrando por Promo
ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=raw_data, color = 'green', hue = 'Promo', style = 'Promo')
```

```python
# Adicionando o tamanho/quantidade
ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=raw_data, color = 'green', hue = 'Promo', size = 'Revenue',
                    sizes=(20, 200))
```

## lmPlots - Gráfico Dispersão com Linha Regressão

```python
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, height=9)
```

```python
# Implot com quebra por Promo
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, hue = 'Promo', ci= False, height=9, markers=["o", "x", "+"])
```

```python
# Visão segregada da quebra
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, col = 'Promo', ci= False, height=5, 
                line_kws={'color': 'red'}, 
                scatter_kws={'color':'#ADC11E'})
```

```python
# Quebra por dia
ax = sns.lmplot(x="Visitors", y="Revenue", data=raw_data, col = 'Day_Name', ci= False, height=4, 
                line_kws={'color': '#031722'}, 
                scatter_kws={'color':'#1E84C1'},
               col_wrap=3,
               x_jitter=.3)
```

## Pairplots - Combinação de gráficos entre features

```python
g = sns.pairplot(raw_data[['Revenue','Visitors','Marketing Spend']], plot_kws={'color':'#0EDCA9'})
```

```python
# Adicionando uma quebra - Promo
g = sns.pairplot(raw_data[['Revenue','Visitors','Marketing Spend', 'Promo']], hue = 'Promo')
```
