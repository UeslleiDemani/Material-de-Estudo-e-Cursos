## Dataset

Nesta prática guiada, usaremos os dados gerados.

## Criação de gráficos
 
Para todos os gráficos do matplotlib, começamos criando uma figura e seus eixos.

### `plt.Figure`

A figura, uma instância da classe plt.Figure, pode ser pensada como um contêiner para todos os objetos que representam eixos, gráficos, textos e rótulos.

- [`matplotlib.figure.Figure()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure)

Para criar uma figura (instância de `plt.Figure`) usamos o método `plt.figure()`

- [`matplotlib.pyplot.figure()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.figure.html)

### `plt.Axes`

Os eixos (axes), uma instância da classe plt.Axes, são formados por uma caixa delimitadora e, eventualmente, conterão o elemento plot com nossa visualização, ticks e rótulos.

- [`matplotlib.axes()`](https://matplotlib.org/3.2.1/api/axes_api.html#matplotlib.axes.Axes)

Para criar um objeto axes (instância de `plt.Axes`) usamos o método `plt.axes()`. 

Uma vez que criamos o objeto axes, usamos o método `ax.plot` para representar graficamente os dados.

- [`matplotlib.axes.Axes.plot()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot)

### Ajustes de cor e estilos

O método `plot` possui argumetnos adicionais que podem ser utilizados para especificar cor e estilo da linha.

#### Alternativas para especificar a cor:

`color = 'blue'` especificamos o nome da cor

`color = 'g'` código de cor abreviado (rgb/cmyk)

`color = '0.75'` escala de cinza entre 0 e 1

`color = '#FFDD44'` código hexadecimal (RGB), de 00 até FF

`color = (1.0, 0.2, 0.3)` tupla RGB com valores entre 0 e 1

#### Alternativas para especificar o estilo da linha:

`linestyle = 'solid'` ou `linestyle = '-'`  linha contínua

`linestyle = 'dashed'` ou `linestyle = '--'` linha tracejada

`linestyle = 'dashdot'` ou `linestyle = '-.'` linha composta (ponto e segmento)

`linestyle = 'dotted'` ou `linestyle = ':'` linha pontilhada

#### Combinações para especificar o estilo e a cor da linha simultaneamente:

`'-g'` linha verde sólida

`'--c'` linha tracejada cor ciano

`'-.k'` linha composta preta

`':r'` linha pontilhada vermelha

A seguir, vamos testar diferentes atributos no mesmo gráfico. Para não reescrever o código de criação desse gráfico cada vez que testamos um novo atributo, vamos definir uma função que retorna esse gráfico "base" que vamos reutilizar.

```python
def base_matplotlib_example():

    # cria a figura
    fig = plt.figure()

    # cria o eixo
    ax = plt.axes()

    # insere os dados no eixo:
    ax.plot(data_x, data_sin, color = '0.75', linestyle = 'dashed', label = 'sin');

    # insere outro gráfico no eixo:
    ax.plot(data_x, data_cos, color = (1.0, 0.2, 0.3), linestyle = 'dotted', label = 'cos');

    # insere outro gráfico no eixo:
    ax.plot(data_x, (data_cos+1), '-g', label = 'cos modified')
    return fig, ax
```

Vamos ver como criar uma figura, chamando a função que definimos:

```python
fig, ax = base_matplotlib_example()
```

### Ajustes dos eixos

Podemos definir os valores mínimo e máximo que serão expressos nos eixos x e y.


Eixo X: `set_xlim`

- [`matplotlib.axes.Axes.set_xlim()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.axes.Axes.set_xlim.html#matplotlib.axes.Axes.set_xlim)

Eixo Y: `set_ylim`

- [`matplotlib.axes.Axes.set_ylim()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.axes.Axes.set_ylim.html#matplotlib.axes.Axes.set_ylim)

```python
fig1, ax1 = base_matplotlib_example()

# muda o eixo x para o range -1 a 11
ax1.set_xlim(-1, 11)    

# muda o eixo y para o range -1.5 a 2.5
ax1.set_ylim(-1.5, 2.5); 
```

Podemos definir os valores dos eixos de modo a se ajustarem aos dados com 'tight':

- [`matplotlib.axes.Axes.axis()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.axis.html#matplotlib.axes.Axes.axis)

```python
fig2, ax2 = base_matplotlib_example()
ax2.axis = 'tight'
```

## Scatter Plots

Os gráficos de dispersão (scatter plots) são utilizados para representar o relacionamento entre duas variáveis numéricas. São semelhantes aos gráficos de linha, mas cada ponto é representado individualmente por um símbolo, em vez de ser unido por segmentos de reta.

### `scatter`

O argumento `marker` define o estilo do símbolo que cada ponto representa. Pode ser uma instância da classe MarkerStyle ou uma string que representa um símbolo específico.

- [`matplotlib.axes.Axes.scatter()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.axes.Axes.scatter.html)

- [`matplotlib.markers.MarkerStyle()`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.markers.MarkerStyle.html)

- [`matplotlib.markers()`](https://matplotlib.org/stable/api/markers_api.html)


O argumento `alpha` define o nível de transparência dos marcadores. É um valor numérico entre 0 (transparente) e 1 (opaco).

```python
fig = plt.figure()

# cria o eixo:
ax = plt.axes()

# insere dados no eixo:
ax.scatter(data_x, data_sin, marker = 'o', color = "green", label = 'serie sin', alpha = 0.4)

# insere mais dados no eixo:
ax.scatter(data_x, data_cos, marker = '+', color = (1.0, 0.2, 0.3), label = 'serie cos', alpha = 0.7);

# insere mais dados no eixo:
ax.scatter(data_x, (data_cos + 1), marker = '*', color = "blue", label = 'serie cos modif', alpha = 1);
```

## Histogramas

### `hist`

- [`matplotlib.axes.Axes.hist()`](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.hist.html)

Entre os parâmetros que podemos definir para um histograma estão:

`bins`: inteiro, lista ou string. 

* Se for um inteiro, define o número de bins (barras) de igual largura na faixa de valores do gráfico.

* Se for uma lista, define os limites de cada barra incluindo o limite esquerdo do primeiro bin e o limite direito do último bin. Nesse caso, eles podem ter larguras diferentes.

* Se for uma string, representa uma das estratégias que `numpy.histogram_bin_edges` suporta, quais sejam: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', 'sqrt'.

`density`: Se for True, os pesos serão normalizados de forma que a integral da função densidade no intervalo seja igual a 1.

`histtype`: string. Tipo de histograma. Os valores possíveis são {'bar', 'barstacked', 'step', 'stepfilled'}.

* 'bar' representa um histograma tradicional

* 'barstacked' representa um histograma de barras empilhadas

* 'step' gera um gráfico (histograma) de linha que por padrão não tem preenchimento

* 'stepfilled' gera um gráfico de linha que por padrão tem preenchimento

`color`: cor das barras do histograma

`edgecolor`: cor das bordas das barras do histograma

`alpha`: valor decimal entre 0 e 1 que representa o nível de transparência, 0 é transparente e 1 é opaco


```python
fig = plt.figure()

# cria o eixo
ax = plt.axes()

# cria o histograma 1
ax.hist(data_normal_1, label = 'data normal 1', alpha = 0.4)

# cria o histograma 2
ax.hist(data_normal_2, label = 'data normal 2', alpha = 0.7);

# cria o histograma 3
ax.hist(data_normal_3, label = 'data normal 3', alpha = 1)
```

Se vamos definir os mesmos valores (alpha, density, e bins, por exemplo)  ao definir os histogramas, é conveniente definir um dicionário com esses valores, e passá-lo como parâmetro em todas as chamadas ao método hist, em vez de definir cada um destes valores para cada um dos histogramas.

Vamos definir os valores de histtype, alpha, density e bins em um dicionário e, em seguida, passar estes dicionário como argumento nos histogramas criados.

```python
kwargs = { 'histtype': 'stepfilled', 
          'alpha': 0.3, 
          'density': False, 
          'bins': 40}

fig = plt.figure()
# cria o eixo
ax = plt.axes()
# cria o histograma 1
ax.hist(data_normal_1, label = 'data normal 1', **kwargs)
# cria o histograma 2
ax.hist(data_normal_2, label = 'data normal 2', **kwargs)
# cria o histograma 3
ax.hist(data_normal_3, label = 'data normal 3', **kwargs);
```

## rcParams

Como mencionamos na parte de estilos, um arquivo rc contém a configuração do matplotlib lida quando ele é iniciado.

Cada vez que iniciamos o matplotlib, uma configuração de tempo de execução (rc) é definida, esta contém os estilos padrões para cada elemento de gráfico que criamos. Podemos ajustar essa configuração modificando os valores de `plt.rc`

- [`matplotlib.pyplot.rc`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.rc.html)

A seguir, veremos um exemplo de como modificar os valores de `rc`. Antes de começar, fazemos uma cópia dos valores rcParams que salvamos na variável `rc_default`, para que possamos restaurá-los posteriormente.

```python
# fazemos um backup dos valores atuais de rc
rc_default = plt.rcParams.copy()

# mudamos alguns valores
from matplotlib import cycler
colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor = '#E6E6E6', edgecolor = 'none',
       axisbelow = True, grid = True, prop_cycle = colors)
plt.rc('grid', color = 'w', linestyle = 'solid')
plt.rc('xtick', direction = 'out', color = 'gray')
plt.rc('ytick', direction = 'out', color = 'gray')
plt.rc('patch', edgecolor = '#E6E6E6')
plt.rc('lines', linewidth = 2)
```

Plotamos um histograma com as configurações correspondentes aos valores `rcParams` modificados.

```python
fig_rcModified = plt.figure()
ax_rcModified = plt.axes()

ax_rcModified.hist(data_normal_1);
```

Restauramos os valores originais de rcParams que salvamos na variável rc_default:

```python
# Removemos alguns dos parâmetros rc_default porque eles serão interrompidos no Matplotlib 3.3 e se forem, ele imprime um aviso:

to_remove = ["animation.avconv_path", 
             "animation.avconv_args", 
             "animation.html_args",
            "keymap.all_axes", 
             "savefig.jpeg_quality", 
             "text.latex.preview"
            ]

for to_remove_key in to_remove:
    if to_remove_key in rc_default.keys():
        print(rc_default[to_remove_key])
        del rc_default[to_remove_key]    
```

```python
# outra maneira de restaurar os valores originais
plt.rcParams.update(rc_default);
```

```python
fig_restore = plt.figure()
ax_restore = plt.axes()
# Se quisermos restaurar os valores de rcParams para o padrão de matplotlib, usamos
ax_restore.hist(data_normal_1);
```

Se quisermos restaurar os valores de rcParams para o padrão de matplotlib, usamos:

```python
# restaura o padrão: 
mpl.rcdefaults()
```
