### Convertendo tipo object para float - `SALE PRICE`.

### 'For' para converter as demais que considerar necessrias:

````python
fields_objects = ['LAND SQUARE FEET','GROSS SQUARE FEET']

for column in fields_objects:
    nyc_sales[column] = pd.to_numeric(nyc_sales[column],downcast='float', errors='coerce')
````

# Verificando varincia das colunas numricas

Funo Auxiliar, seleo de tipos
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.select_dtypes.html

````python
# busca todos os numricos possveis                                      
variaveis_numericas_nyc = nyc_sales.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64'])
````

````python
variaveis_numericas_nyc
````

````python
# aplicando escalonamento - default=(0, 1)
variaveis_numericas_escaladas_nyc = minmax.fit_transform(variaveis_numericas_nyc) 
````

````python
variaveis_numericas_escaladas_nyc
````

````python
#armazenando as variveis j tranformadas em um  novo dataframe
variaveis_numericas_escaladas_nyc = pd.DataFrame(variaveis_numericas_escaladas_nyc, columns = variaveis_numericas_nyc.columns)
````

````python
variaveis_numericas_escaladas_nyc.head()
````

### Calculando varincia com mtodo np.var() - numpy
- https://numpy.org/doc/stable/reference/generated/numpy.var.html


No queremos varincia muito prximas de zero, ento vamos filtrar variancia > 0.001

````python
variaveis_numericas_escaladas_nyc['LAND SQUARE FEET']
````

````python
np.var(variaveis_numericas_escaladas_nyc['LAND SQUARE FEET'])
````

````python
#verificando varincia da coluna numrica 'LAND SQUARE FEET'
np.var(variaveis_numericas_escaladas_nyc['LAND SQUARE FEET']) >0.001  
````

````python
#verificando varincia da coluna numrica ''COMMERCIAL UNITS'

np.var(variaveis_numericas_escaladas_nyc['COMMERCIAL UNITS'])  >0.001
````

## Melhor usar for para percorrer as colunas numricas

````python
listLowVariance = []
for col in variaveis_numericas_escaladas_nyc.columns:
    if np.var(variaveis_numericas_escaladas_nyc[col]) > 0.001:
        print(f'{col} possui varincia considervel')
    else:
        print(f'{col} possui baixa varincia')
        listLowVariance.append(col)
````

````python
listLowVariance
````
